# Bar Graphs

```python
import matplotlib.pyplot as plt
```

# 1. Simple Bar Graph

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.bar(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Sachin Tendulkar's yearly runs")
```

# 2. Side by Side Graph

```python
import numpy as np
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
x = np.arange(len(years))     #Years Indices
width = 0.25

plt.bar(x-width, sachin, width=width, label="Sachin Tendulkar")
plt.bar(x, sehwag, width=width, label="Virender Sehwag")
plt.bar(x+width, kohli, width=width, label="Virat Kohli")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs Comparison")
plt.legend()
plt.xticks(x,years)    #to show years instead of 1,2,3...
plt.tight_layout
plt.show()
```

# 3. Horizontal Bar Graph

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100,
300+600+800+1100+900]
plt.barh(players, runs_5yrs, color="skyblue")
plt.xlabel("Runs Scored in 5 Years")
plt.ylabel("Players")
plt.title("Total Runs Scored in 5 Years")
plt.tight_layout()
plt.show()
```

# 4. Adding Value Labels with plt.text()
---
**You can use plt.text() to display values on top of bars, making your chart easier to read. It's especially
useful in presentations and reports.**

```python
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]
plt.bar(players, runs, color="black")
for i in range(len(players)):
    plt.text(i, runs[i] + 30, str(runs[i]), ha="center")
plt.xlabel("Players")
plt.ylabel("Runs Scored")
plt.title("Runs Scored by Players")
plt.tight_layout
plt.show()
```
