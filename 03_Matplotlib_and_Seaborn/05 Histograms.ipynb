# Histograms
***
**A histogram is a type of plot that shows the distribution of a dataset. It’s especially useful for visualizing the
frequency of numerical data within specified ranges (called bins).**

# Why and When to Use Histograms?

Use Histograms When:

1. You want to understand the distribution of a numerical dataset (e.g. age, salary, views).
2. You want to detect skewness, outliers, or understand spread.
3. You’re binning continuous data into intervals

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
marks = [
  4, 6, 8, 10, 50, 95, 45, 72, 18, 20,
  22, 24, 25, 26, 28, 30, 32, 34, 35, 36,
  38, 40, 42, 44, 45, 46, 48, 50, 52, 54,
  56, 58, 60, 62, 64, 66, 68, 70, 72, 74,
  70, 75, 80, 85, 88, 90, 92, 94, 96, 98
]
plt.hist(marks, bins=5, edgecolor="black")
plt.xlabel("Marks")
plt.ylabel("Frequency of Marks")
plt.title("Marks Scored by Students")
plt.show()
```

# Example - Age Distribution of Youtube Viewers

```python
ages = [
 34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
 25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
 44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
 23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
 37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
 33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
 25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
 33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
 27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
 30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]
bins = [10,20,30,40,50,60,70]
plt.hist(ages, bins=bins, edgecolor="black")
plt.axvline(np.mean(ages), color="red", linestyle="--", linewidth="2", label="Average Age")
plt.axvline(np.median(ages), color="black", linestyle="-", linewidth="2", label="Median Age")
plt.xlabel("Ages")
plt.ylabel("No of Viewers")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.show()
```
