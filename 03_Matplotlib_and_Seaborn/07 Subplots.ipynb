# Subplots
---
Subplots allow you to show multiple plots in a single figure, side-by-side or in a grid layout. This is helpful
when comparing different datasets or aspects of the same data

```python
import matplotlib.pyplot as plt
```

# Basic Subplot

```python
x = [1,2,3,4,5,6,7,8]
y1 = [i*2 for i in x]
y2 = [i**2 for i in x]

plt.subplot(1,2,1)        # (rows, cols, plot_no)
plt.plot(x, y1)
plt.title("Double of x")

plt.subplot(1,2,2)       # (rows, cols, plot_no)
plt.plot(x,y2)
plt.title("Square of x")

plt.tight_layout()
plt.show()
```

# 2 X 2 Grid Subplots

```python
x = [1,2,3,4,5,6,7,8]
y1 = [i*2 for i in x]
y2 = [i**2 for i in x]
y3 = [i**0.5 for i in x]
y4 = [10-i for i in x]

plt.subplot(2,2,1)        # (rows, cols, plot_no)
plt.plot(x, y1)
plt.title("Double of x")

plt.subplot(2,2,2)       # (rows, cols, plot_no)
plt.plot(x,y2)
plt.title("Square of x")

plt.subplot(2,2,3)        # (rows, cols, plot_no)
plt.plot(x, y3)
plt.title("square root of x")

plt.subplot(2,2,4)       # (rows, cols, plot_no)
plt.plot(x,y4)
plt.title("10-x")

plt.tight_layout()
plt.show()
```

# Using plt.subplots() for Clean Code 

```python
x = [1,2,3,4,5,6,7,8]
y1 = [i*2 for i in x]
y2 = [i**2 for i in x]
fig, axs = plt.subplots(1,2,figsize=(10,4))
axs[0].plot(x,y1)
axs[0].set_title("x*2")

axs[1].plot(x,y2)
axs[1].set_title("x**2")

fig.suptitle("Simple Comparison Plots", fontsize=14)
fig.savefig("my_plots.png")
plt.tight_layout()
plt.show()
```

# Looping Over the Subplots

```python
x = [1,2,3,4,5,6,7,8]
y1 = [i*2 for i in x]
y2 = [i**2 for i in x]
y3 = [i**0.5 for i in x]
y4 = [10-i for i in x]

fig, axs = plt.subplots(2,2,figsize=(8,6))
ys = [y1,y2,y3,y4]
titles = ["x*2", "squared x", "sqrt(x)", "(10-x)"]

for i in range(2):
    for j in range(2):
        idx = i*2+j
        axs[i,j].plot(x, ys[idx])
        axs[i,j].set_title(titles[idx])

fig.suptitle("Comparison of X with Different Operation")
plt.tight_layout()
plt.show()
```
