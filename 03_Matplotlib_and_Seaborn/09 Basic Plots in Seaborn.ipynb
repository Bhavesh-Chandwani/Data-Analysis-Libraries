```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# How to see all Available Data Sets

```python
sns.get_dataset_names()
```

# How to Load a Dataset

```python
tips = sns.load_dataset('tips')
tips.head()
```

**Why Seaborn Provides Datasets** :

1. To quickly test different types of plots
2. To learn plotting without needing your own data at first
3. To create examples and tutorials easily
4 .To show real-world messy data handling (missing values, categorical data, etc.)

# Line Plot

```python
sns.set_theme(style="darkgrid")
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title("Line Plot Example")
plt.show()
```

# Scatter Plot

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time")
plt.title("Scatter Plot with color by time")
plt.show()
```

# Bar Plot

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Average Bill Per day")
plt.show()
```

# Box Plot

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill Per day")
plt.show()
```

# HeatMap

```python
flights = sns.load_dataset('flights')
pivot_table = flights.pivot(index="month", columns="year",values="passengers")
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Passengers')
plt.show()
```

# Working with Pandas Data Frame
