**Q. Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same
using a horizontal bar chart and observe which is easier to understand.**

# 1. Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
mobile_os = ["Android", "iOS", "Others"]
market_share = [72, 26, 2]   # in percentage
colors = ["#9EDEA7", "#42D4D4", "#D49EDE"]
plt.figure(figsize=(6,6))
explode = [0.1,0,0]
plt.pie(market_share, labels=mobile_os, colors=colors, wedgeprops={"edgecolor" : "black", "linewidth" : 1}, shadow=True, autopct="%1.1f%%", explode=explode)
plt.title("The Shares of Smart Phones Market")
plt.axis('equal')
plt.show
```

# 2. Horizontal

```python
plt.barh(mobile_os, market_share, color="#D79EDE")
plt.xlabel("Market Share in Percentage")
plt.ylabel("Mobile_Os")
plt.title("The Shares of Smart Phones Market")
plt.tight_layout()
plt.show()
```
