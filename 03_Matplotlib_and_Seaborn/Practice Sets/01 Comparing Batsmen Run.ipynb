```python
import matplotlib.pyplot as plt
```

```python
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
kohli = [1200, 1350, 1450, 1600, 1550, 1700, 1800, 1900, 1850, 1500]
rohit = [900, 1000, 1100, 1200, 1250, 1400, 1500, 1600, 1550, 1300]
dhoni = [1500, 1400, 1300, 1200, 1400, 1000, 900, 500, 600, 500]
plt.plot(years, kohli, color="red", marker="o", linestyle="--", linewidth=2, label="Virat Kohli")
plt.plot(years, rohit, color="magenta", marker="^", linestyle="-", linewidth=2, label="Rohit Sharma")
plt.plot(years, dhoni, color="black", marker="D", linestyle="-.", linewidth=2, label="MS Dhoni")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Batsmen Runs Comparison")
plt.grid(True)
plt.legend()
plt.show()
```
