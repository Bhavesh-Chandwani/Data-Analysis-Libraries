# Real Data Example: 
**Sachin Tendulkarâ€™s Runs Over Time**


```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years, runs)
 # Giving Labels to axes
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Sachin Tendulkars Yearly Runs")
plt.show()
```

# Multiple Lines in One Plot

```python
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years, sachin, label="Sachin Tendulkar")
plt.plot(years, sehwag, label="Virendar Sehwag")
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

# Format Strings

```python
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years, sachin, 'ro--', label="Sachin Tendulkar")
plt.plot(years, sehwag, 'kD:' ,label="Virendar Sehwag")
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

# Colors, Markers and Line Style Arguements

```python
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years, sachin, color = "Magenta", marker = "^", linestyle = "-.", label="Sachin Tendulkar")
plt.plot(years, sehwag, color = "cyan", marker = "*", linestyle = "-" ,label="Virendar Sehwag")
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

# Line Width and Layout Tweaks

```python
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years, sachin, linewidth = 3, label="Sachin Tendulkar")
plt.plot(years, sehwag, linewidth = 2, label="Virendar Sehwag")
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
