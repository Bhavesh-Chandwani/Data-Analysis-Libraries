# Pie Charts
---
**Pie charts are used to show part-to-whole relationships. They're visually appealing but best used with fewer
categories, as too many slices can get cluttered and hard to interpret.**


```python
import matplotlib.pyplot as plt
```

# Setting Style

```python
plt.style.use("ggplot")
```

# Basic Pie Chart

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs, labels=labels)
plt.title("Career Runs of Indian Batsmen")
plt.show()
```

# Custom Colors

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ["#B56A69", "#6CEB6C", "#EB6CEB", "#6CD2EB"]
plt.pie(runs, labels=labels, colors=colors)
plt.title("Career Runs of Indian Batsmen")
plt.show()
```

# Add Edges and Style Slices with **wedgeprops**


```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ["#B56A69", "#6CEB6C", "#EB6CEB", "#6CD2EB"]
plt.pie(runs, labels=labels, colors=colors, wedgeprops={"edgecolor" : "black", "linewidth" : "2", "linestyle" : "--"})
plt.title("Career Runs of Indian Batsmen")
plt.show()
```

# Highlight a Slice Using **explode
***
Use **explode** to pull out one or more slices

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ["#B56A69", "#6CEB6C", "#EB6CEB", "#6CD2EB"]
explode = [0.1,0,0,0]
plt.pie(runs, labels=labels, colors=colors, wedgeprops={"edgecolor" : "black", "linewidth" : "2", "linestyle" : "--"}, explode=explode)
plt.title("Representation of Exploded Slice")
plt.show()
```

# Add Shadows

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ["#B56A69", "#6CEB6C", "#EB6CEB", "#6CD2EB"]
explode = [0.1,0,0,0]
plt.pie(runs, labels=labels, colors=colors, explode=explode, shadow=True)
plt.title("Representation of Exploded Slice")
plt.show()
```

# Start Angle

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ["#B56A69", "#6CEB6C", "#EB6CEB", "#6CD2EB"]
explode = [0.1,0,0,0]
plt.pie(runs, labels=labels, colors=colors, explode=explode, shadow=True, startangle=180)
plt.title("Representation of Exploded Slice")
plt.show()
```

# Auto Percentage
---
To Show Percentage with **autopct**

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ["#B56A69", "#6CEB6C", "#EB6CEB", "#6CD2EB"]
explode = [0.1,0,0,0]
plt.pie(runs, labels=labels, colors=colors, explode=explode, shadow=True, startangle=180, autopct="%1.1f%%")
plt.title("Representation of Exploded Slice")
plt.show()
```
