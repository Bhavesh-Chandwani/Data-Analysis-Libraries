# Scatter PLot
---
Scatter plots are used to show the relationship between two variables. Let's assume we are plotting the study
hours vs. exam scores for a group of students.

```python
import matplotlib.pyplot as plt
```

# Basic Scatter Plot

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 80, 50, 55, 20, 65, 75, 85, 90]
plt.scatter(study_hours, exam_scores)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study Hrs vs Exam Scores")
plt.grid(True)
plt.show()
```

# Adding Color and Size

```python
size = [score*2 for score in exam_scores]
color = ["red" if score < 60 else "green" for score in exam_scores]
plt.scatter(study_hours, exam_scores, s=size, c=color)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Color and Sized Scatter Plot")
plt.grid(True)
plt.show()
```

# Using a Color Map
---
**cmap adds gradient coloring based on the score. colorbar helps understand what the colors
represent.**

```python
plt.scatter(study_hours, exam_scores, c=exam_scores, cmap="plasma")
plt.colorbar(label="Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam SCores")
plt.title("Scatter Plot with Color Map")
plt.grid(True)
plt.show()
```

# Adding Annotations

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 80, 50, 55, 20, 65, 75, 85, 90]
plt.scatter(study_hours, exam_scores, c=exam_scores, cmap="plasma")
plt.colorbar(label="Scores")
for i in range (len(study_hours)):
    plt.annotate(f"S{i+1}", (study_hours[i], exam_scores[i]-3))
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.title("Sactter Plots with Annotation")
plt.grid(True)
plt.show()
```

# Multiple Groups in One Plot

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 90, 95, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label="A", color="red")
plt.scatter(class_b_hours, class_b_scores, label="B", color="yellow")

plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.title("Scatter Plot : Class A v/s Class B")
plt.grid(True)
plt.legend()
plt.show()
```
