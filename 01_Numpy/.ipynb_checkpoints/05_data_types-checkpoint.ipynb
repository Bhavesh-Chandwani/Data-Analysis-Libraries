```python
import numpy as np
```

```python
# Checking Data Type
arr = np.array([1,2,3,4,5])
print(arr.dtype)       # int 64 or int 32 depending on the system architecture.
print(arr.nbytes)
```

**1. Changing Data Type**

```python
# Changing Data Type using "astype"
arr_float = arr.astype("float32")
print(arr_float, arr_float.dtype)

#We can define Data Type while creatiing  variable
arr2 = np.array([10,20,30,40], dtype="float64")
print(arr2, arr2.dtype)
```

**2. Downcasting Data Type To Save Memory.**

```python
arr_large = np.array([100000, 200000, 300000], dtype=np.int64)
arr_small = arr_large.astype(np.int32)
print(arr_small)
print(arr_small.dtype)
print(arr_small.nbytes)
```

**3. Complex Numbers**

```python
arr = np.array([1+2j, 3+4j], dtype = np.complex64)
print(arr.real)   # Printing the real part 
print(arr.imag)   # Printing the imaginary part
print(arr.dtype)
```

**4. Object Data Type** - If you need to store mixed or complex data types (e.g., Python objects), you can
use dtype='object' . However, this type sacrifices performance, so it should only
be used when absolutely necessary

```python
arr = np.array([{'a': 1}, [1, 2, 3], 'hello'], dtype=object)
print(arr.dtype)
```
