**3. Broadcasting: Scaling Arrays Without Extra Memory**

```python
arr2 = np.array([10,20,30,40,50])
result = arr2 + 10  #Broadcasting : 10 added to all elements
print(result)
```

**4. Broadcasting of Array with Different Shapes**

```python
# Broadcasting with Two Arrays
arr1 = np.array([10,20,30])
arr2 = np.array([1,2,3])
sum = arr1 + arr2
print(sum)
```

```python
# Broadcasting a 2D Array and a 1D Array
arr1 = np.array([[10,20,30], [40,50,60]])
arr2 = np.array([1,2,3])
sum = arr1 + arr2
print(sum)
```

**5. Hands-on: Applying Broadcasting to Real-World Scenarios**

```python
# Simulating a dataset (5 samples, 3 features)
data = np.array([[10, 20, 30],
[15, 25, 35],
[20, 30, 40],
[25, 35, 45],
[30, 40, 50]])
# Calculating mean and standard deviation for each feature (column)
mean = data.mean(axis=0)
std = data.std(axis=0)

# Normalizing the data using the broadcast
normalize_data = (data-mean)/std
print(normalize_data)
```
