```python
# Calculating time of 2 lists
import numpy as np
import time
l1 = [10, 13, 15, 85, 69]
l2 = [11, 15, 18, 90, 79]

start = time.time()
multiplication = [x*y for x,y in zip(l1, l2)]
end = time.time()
print(multiplication)
print(end-start)
```

```python
# Calculating time of 2 numpy arrays
import numpy as np
import time
arr1 = np.array([10, 13, 15, 85, 69])
arr2 = np.array([11, 15, 18,  90, 79])

start = time.time()
multiplication = arr1 * arr2
end = time.time()
print(multiplication)
print(end-start)
```
