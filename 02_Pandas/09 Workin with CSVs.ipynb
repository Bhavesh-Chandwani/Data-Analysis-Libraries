# Reading & Writing Files in Pandas


# CSVs Files
---
**Read CSV**

```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df
```

```python
df = df[df['CGPA'] > 8.5]
df
```

**Write CSV**

```python
df.to_csv("updated_data.csv", index=False)   #Index will not be highlighted in excel
```

**Creating Excel File**

```python

data = {
    "Region": ["East", "West", "East", "West"],
    "Month": ["Jan", "Jan", "Feb", "Feb"],
    "Revenue": [12000, 15000, 13000, 16000]
}
df = pd.DataFrame(data)
df
```

```python
df.to_excel("Sales_Data.xlsx", sheet_name="Sales", index=False)
```

**Reading Excel File**

```python
df_read = pd.read_excel("Sales_Data.xlsx", sheet_name="Sales")
df_read
```

```python
# Summarize revenue by region
summary = df_read.groupby("Region")["Revenue"].sum().reset_index()
summary
```

**Write Excel**

```python
summary.to_excel("region_summary.xlsx", index=False)
```
