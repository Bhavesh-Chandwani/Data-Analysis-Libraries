**Data Loading**

```python
import pandas as pd
```

```python
df = pd.read_csv("MCU Movies Data.csv")
df
```

**1. Selecting Rows and Column**

**i. Selecting Columns**

```python
print(df['Actor(s)']) # Access Actors rows, Single column (as Series)
print(type(df['Actor(s)']))    # Checking the type which is ofcourse "Series"
```

```python
df['Actor(s)'][3]    #Accessing Actors name through index
```

```python
df[['Actor(s)', 'IMDb Rating']]     #Accessing Actors column and IMDb colums as Data Frame (Multiple Column)
```

**2.Selecting Rows by Index**

**i.Use .loc[] (label-based)** :- Using loc[] when the rows have specific name like "a,b,c,d" so we can access the
rows like df.loc['a','Film']  # Avengers: Endgame. But here we have indexes so we will use index for now.

```python
df.loc[0]      # First row (by label)
```

**ii Use .iloc[] (position-based)**: We can access rows and column through their indexes irrespective of their labels


```python
df.iloc[0]
```

**iii. Select Specific Rows and Columns**

```python
df.loc[0, 'Actor(s)']
```

```python
df.iloc[0,1]
```

**iv. Slicing**

```python
df.loc[0:2, ['Film', 'Actor(s)']]     # Here the "2" is included because it is sliced according to labels (.loc)
```

```python
df.iloc[0:2, 0:2]
```

**v. Fast Access .at and .iat** : These are optimized for single element access. Same logic as **".loc[]"** and **.iloc[]**

```python
df.at[0, 'Year']
```

```python
df.iat[0, 2]
```

**2. Filtering with Conditions**

**i. Simple Conditions.**

```python
df[df['IMDb Rating'] > 7]
```

```python
df[df['Year'] > 2020]
```

```python
df[df['Box Office Collection'] > "$760M"]
```

**ii. Multiple Conditions**

```python
df[df['IMDb Rating'] > 8]['Actor(s)']
```

```python
df[(df['IMDb Rating'] > 7) | (df['Year'] > 2018)]
```

**3. Querying with .query()** - The .query() method in pandas lets you filter DataFrame rows using a string
expression — it’s a more readable and often more concise alternative to using
boolean indexing.
This is a cleaner, SQL-like way to filter:

```python
df.query("`IMDb Rating` > 7 and Year > 2020")
```

**i. Dynamic Column Names**

```python
col = "Year"
df.query(f"{col} > 2018")
```
