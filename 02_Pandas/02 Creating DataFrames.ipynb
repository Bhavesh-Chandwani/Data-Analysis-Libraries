```python
import pandas as pd
```

**1. From Python Lists**

```python
data = [
["Alice", 25],
["Bob", 30],
["Charlie", 35],
["Michael", 24]
]

df = pd.DataFrame(data, columns=["Name", "Age"])
df
```

**2. From DIctionary of List**

```python
data = {
"Name": ["Alice", "Bob", "Charlie", "Michael"],
"Age": [25, 30, 35, 24]
}
df = pd.DataFrame(data)
df
```

**3. From Numpy Arrays**

```python
import numpy as np
data = np.array([[1,2,3], [4,5,6]])
df = pd.DataFrame(data, columns=["A", "B", "C"])
df
```

**4. Read Excel Files**

```python
df = pd.read_excel("data.xlsx")
df
```

**5. Read csv files.**

```python
df = pd.read_csv("data.csv")
df
```

**6. From Json File**

```python
df = pd.read_json("data.json")
df
```

**7. Read CSV from URL**

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
df
```
